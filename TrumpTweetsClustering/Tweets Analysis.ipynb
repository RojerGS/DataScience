```python
# General:
import pandas as pd     # To handle data
import numpy as np      # For number computing
import json
import nltk
nltk.download('stopwords')
import re as re
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

# Create a pandas dataframe
with open(r'C:\Users\Dieuwke\Documents\MSc\MSc AM\Mathematical Data Science\Trumps Tweet Analysis Project\DataScience\TrumpTweetsClustering\Data\TrumpTweets2018.json') as json_file:
    jdata = json.load(json_file)
    
data = pd.DataFrame(data=jdata)

data.drop(["created_at", "favorite_count", "id_str", "retweet_count", "source"], axis=1, inplace=True)
data.columns = ["is_retweet","text"]
data = data[data.is_retweet == False]
data.drop(["is_retweet"], axis=1, inplace=True)

# #Create data file with original tweets (without retweets)
# data_original = data.reset_index()
# data_original.drop(["index"], axis = 1, inplace=True)
# data_original.to_csv('data_original.csv')
# display(data_original.head(100))

#Cleaning of tweets

from nltk.corpus import stopwords
stop = stopwords.words('english')

data["text"] = data["text"].str.replace("(@[A-Za-z]+)|([^A-Za-z \t])|(\w+:\/\/\S+)"," ")
data["text"] = data["text"].str.replace("amp"," ")
data["text"] = data["text"].str.lower().str.split()
data["text"] = data["text"].apply(lambda x: ' '.join([word for word in x if word not in stop]))
data["text"] = data["text"].str.split()
display(data.head(10))
data["text"] = data["text"].apply(lambda x: [stemmer.stem(y) for y in x])
data = data.reset_index()
data.drop(["index"], axis = 1, inplace=True)
display(data.head(100))

data.to_csv('data_clean_stemmed_withoutRT.csv')
```
